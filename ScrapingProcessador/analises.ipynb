```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb

df = pd.read_excel('./processadores_filtrados.xlsx')
```

```python
plt.figure(figsize=(6, 6))
proporcao_marcas = df['marca'].value_counts()
proporcao_marcas.plot(kind='pie', autopct='%1.1f%%', colors=['#1f77b4', '#ff7f0e'],
    textprops={'fontsize': 12, 'fontweight': 'bold'})

plt.title('Distribuição marcas')
plt.ylabel('')
plt.legend()

plt.show()
```

```python
top_processadores = df.sort_values(by='preco_atual', ascending=False).head(5)

plt.figure(figsize=(10,5))

valor_processador = top_processadores.groupby('modelo')['preco_atual'].sum()
valor_processador.plot(kind='bar', color='#1f77b4')

plt.title('Processadores mais caros no momento')

plt.xlabel('Processadores')
plt.ylabel('Preco Atual')


plt.grid(True)
plt.xticks(rotation=360)
plt.legend()

# Mostrar o gráficos
plt.show()
```

```python
plt.figure(figsize=(10,5))

quantidade_estoque = df.groupby('marca')['em_estoque'].sum()
quantidade_estoque.plot(kind='bar', color='purple')

plt.title('Quantidade em estoque por marca')

plt.xlabel('Marcas')
plt.ylabel('Em Estoque')


plt.grid(True)
plt.xticks(rotation=360)
plt.legend()

# Mostrar o gráficos
plt.show()
```

```python
plt.figure(figsize=(10,5))

valor_processador = df.groupby('preco_atual')['freq_base_GHz']
valor_processador.plot(marker='s', color='#1f77b4')

plt.title('Processadores mais caros no momento')

plt.xlabel('Processadores')
plt.ylabel('Preco Atual')


plt.grid(True)
plt.xticks(rotation=360)
plt.legend()

# Mostrar o gráficos
plt.show()
```
